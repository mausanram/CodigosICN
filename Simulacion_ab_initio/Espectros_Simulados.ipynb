```python
#### Visualizar las keys del diccionario ####
list_path = path.split('_')
print('Las llaves del diccionairo son: ', data_histogram.keys())
list_thet = data_histogram['Theta_Deg']
list_phi = data_histogram['Phi_Deg']
list_energy_SM = data_histogram['Energy_SM']
list_energy_DP = data_histogram['Energy_DP']
```

```python
fig, axs = plt.subplots(figsize = [10,7])
axs.hist(list_thet, bins=30,  histtype = 'step', color = 'b', label = 'Muones simulados: ' + str(2000))

axs.set_xlim(0, 90)
axs.set_xlabel('Ángulo (°)')
axs.legend()
fig.suptitle(r'Distribución angular $\theta$')

plt.show()
```

```python
fig, axs = plt.subplots(figsize = [10,7])
axs.hist(list_phi, bins=15,  histtype = 'step', color = 'b', label = 'Muones simulados: ' + str(2000))

# axs.set_xlim(0, 180)
axs.set_xlabel('Ángulo (°)')
axs.legend()
fig.suptitle(r'Distribución angular $\phi$')

plt.show()
```

```python
fig, axs = plt.subplots(figsize = [10,7])
axs.hist(list_energy_SM,  histtype = 'step', color = 'b', bins=100, label = 'Muones simulados: ' + str(1000), range = [0, 30000])

# axs.set_xlim(0, 20000)
# axs.set_ylim(0, 200)
axs.set_xlabel('Energía (MeV)')
axs.legend()
# axs.set_yscale('log')
# axs.set_xscale('log')
fig.suptitle(r'Distribución de energías de Smith-Duller')

plt.show()
```
